#**Drop Unnamed Column**




```python
# This line not run twice if you alredy runed because its droped the unnamed column
df = df.drop("Unnamed: 0", axis=1)
```

```python
df.head(20)
```

# **Change weekly study hours column**

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("05-Oct","5-10")
```

```python
df.head(20)
```

# **gender distribution**

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#From the above chart we have analysed that:
#The number of females in the data is more than the number of males
```

```python
gp = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gp
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gp, annot = True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

```python
gp1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gp1
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gp1, annot = True)
plt.title("Relationship between Parent's marital Status and Student Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/nagligible impact
#on the student' s score due to there parent marital status
```

```python
gp2 = df.groupby("PracticeSport").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
gp2
```

```python
plt.figure(figsize = (4,4))
sns.heatmap(gp2, annot = True)
plt.title("Relationship between Sports Practice and Student Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/nagligible impact
#on the student' s score due to there sports practice
```

# **Distribution of ethnic group**

```python
plt.figure(figsize = (4,4))
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l = ["group A","group B","group C","group D","Group E"]
mylist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mylist)
plt.pie(mylist, labels = l, autopct = "%1.2f%%" )
plt.title("Distribution of Ethnic Groups")
plt.show()
```
